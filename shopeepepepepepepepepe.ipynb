```python
import requests

#Go to dev tools, select network, filter to XHR, review requests (usually the largest), get header info
#modify this however you need to query the api e.g. sorting, page (experiment on the website first!)
url = 'https://shopee.ph/api/v4/search/search_items?by=relevancy&limit=60&match_id=11021036&newest=0&order=desc&page_type=search&scenario=PAGE_CATEGORY&version=2'

#copy relevant header info
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36',
    'X-Requested-With': 'XMLHttpRequest',
    'Referer': 'https://shopee.ph/Makeup-Fragrances-cat.11021036',
}    

r = requests.get(url, headers=headers)

# this is the "page" data 
data = r.json()
```
