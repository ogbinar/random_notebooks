```python
import requests
from bs4 import BeautifulSoup

def scrape_blog(url):
    # Make a request to the website
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find all article titles (adjust the selector based on the structure of the website)
        article_titles = soup.find_all('h3', class_='post-title entry-title')

        # Print the titles
        for title in article_titles:
            print(title.text)
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
target_url = "https://cofibean.blogspot.com/"

scrape_blog(target_url)
```

```python
for i in soup2.find_all("div", class_='country-info'):
    p = i.find_all('span')
    for idx,j in enumerate(i.find_all('strong')):
        print(j.text,p[idx].text)
```

```python
import pandas as pd

capital_list = []
for i in soup2.find_all("div", class_='country-info'):
    
    capital_dict = {}
    p = i.find_all('span')
    for idx,j in enumerate(i.find_all('strong')):
        capital_dict[j.text] = p[idx].text
    capital_list.append(capital_dict)

capital_df = pd.DataFrame(capital_list)
```
