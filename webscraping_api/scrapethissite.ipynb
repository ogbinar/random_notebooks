```python
url = "https://www.scrapethissite.com/pages/forms/?page_num=24"
r = requests.get(url)
```

```python
html_doc = r.text
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
teams = soup.find_all("tr",class_="team")
```

```python
len(teams)
```

```python
teams[0]
```

```python
teams[0].find_all("td")[0].text.strip()
```

```python
team_data = []
for i in teams:
  name = i.find_all("td")[0].text.strip()
  year = i.find_all("td")[1].text.strip()
  wins = i.find_all("td")[2].text.strip()
  losses = i.find_all("td")[3].text.strip()

  data_dict = {
      "name" : name,
      "year" : int(year),
      "wins" : int(wins),
      "losses" : int(losses)
  }

  team_data.append(data_dict)
  print(name,year,wins,losses)
```
