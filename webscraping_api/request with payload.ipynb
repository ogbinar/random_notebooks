```python
import requests

url = "https://cmci.dti.gov.ph/data-portal-process.php"

headers = {

    "Origin": "https://cmci.dti.gov.ph"
}

payload = {
 
    'chk-indicators[]': 'intc',
    'chk-lgu[]': 'Aborlan',
    'chk-year[]': [
        '2014', '2015', '2016', '2017', '2018',
        '2019', '2020', '2021', '2022', '2023'
    ]

}

response = requests.post(url, headers=headers, data=payload)

print(response.status_code)
print(response.text)
```

```python
from bs4 import BeautifulSoup

# Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table element based on its ID or class
table = soup.find('table', {'id': 'table_data2'})

# If the table is found, you can extract its content
if table:
    # Extracting table rows
    rows = table.find_all('tr')

    # Loop through rows and print the data
    for row in rows:
        columns = row.find_all(['th', 'td'])
        row_data = [column.get_text(strip=True) for column in columns]
        print(row_data)
else:
    print("Table not found.")
```

```python
rows_list = []
for row in rows:
    columns = row.find_all(['th', 'td'])
    row_data = [column.get_text(strip=True) for column in columns]
    rows_list.append(row_data)
print(rows_list)
```

```python
import pandas as pd

pd.DataFrame(rows_list)
```
