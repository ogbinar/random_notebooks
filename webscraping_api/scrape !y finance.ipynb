```python
# Import necessary libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Make a GET request to the Yahoo Finance gainers page
response = requests.get("https://finance.yahoo.com/gainers")

# Parse the HTML response using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Extract column names from the table headers
colnames = [z.text for z in soup.find_all('th')]

# Initialize an empty list to store rows of data
rows_list = []

# Iterate over each table row with the class "simpTblRow"
for tr in soup.find_all("tr", class_="simpTblRow"):
    # Initialize an empty list to store data for each column in the row
    col_list = []
    
    # Iterate over each table cell in the row
    for td in tr.find_all("td"):
        # Append the text content of each cell to the column list
        col_list.append(td.text)
    
    # Append the column list (representing a row) to the rows list
    rows_list.append(col_list)

# Create a DataFrame from the list of rows and column names
df = pd.DataFrame(rows_list, columns=colnames)

# Display the DataFrame
df
```
