```python
import requests
from bs4 import BeautifulSoup
import json

search_term="mojacko"
#Go to dev tools, select network, filter to XHR or docs, review requests (usually the largest), get header info
#modify this however you need to query the api e.g. sorting, page (experiment on the website first!)
url = 'https://www.lazada.com.ph/catalog/?q='+search_term+'&_keyori=ss&from=input&spm=a2o4l.searchlist.search.go.2f34378fZQRtZd'

#copy relevant header info
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36',
    'Referer': 'https://www.lazada.com.ph/catalog/?q=mojacko&_keyori=ss&from=input&spm=a2o4l.home.search.go.239e359dFlDlaR',
}    

r = requests.get(url,headers=headers)

soup = BeautifulSoup(r.content, 'html.parser')
jsonStr = soup.find_all("script")[2].text.split("\n")[1].split("pageData =")[1].strip().strip(";")
jsonObj = json.loads(jsonStr)
```
